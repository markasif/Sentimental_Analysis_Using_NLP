```python
import re
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
```

```python
# Removing special characters
data["reviewText"] = data["reviewText"].apply(lambda x: re.sub(r'[^a-zA-Z0-9 ]+', '', str(x)))

# Removing stopwords
stop_words = set(stopwords.words("english"))  # Load the stopwords list
data["reviewText"] = data["reviewText"].apply(lambda x: " ".join([word for word in str(x).split() if word.lower() not in stop_words]))

# Removing URLs
data["reviewText"] = data["reviewText"].apply(lambda x: re.sub(r'https?://\S+|www\.\S+', '', str(x)))

# Removing HTML tags
data["reviewText"] = data["reviewText"].apply(lambda x: BeautifulSoup(str(x), 'lxml').get_text())

# Removing additional spaces
data["reviewText"] = data["reviewText"].apply(lambda x: " ".join(str(x).split()))
```

```python
data
```

```python
## using lemmatizer
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
def lemmatize_word(text):
    return ' '.join([lemma.lemmatize(word) for word in text.split()])
```

```python
data["reviewText"]=data["reviewText"].apply(lambda x : lemmatize_word(x))
```

```python
from nltk import word_tokenize
from gensim.utils import simple_preprocess
```

```python
from nltk import word_tokenize

sentences = []  # List of tokenized sentences
for i in range(len(data)):
    text = data["reviewText"][i]
    tokenized_text = word_tokenize(text)  # Tokenize each sentence
    sentences.append(tokenized_text)  # Add tokenized sentence to the list
```

```python
from gensim.models import Word2Vec

# Train Word2Vec on tokenized sentences
model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)

# To get all vocabulary
vocab = model.wv.index_to_key
print(vocab[:10])  # Display the top 10 words in the vocabulary
```

```python
import numpy as np
```

```python
def compute_avg_word2vec(sentence, model, vector_size):
    """
    Compute the average Word2Vec for a single sentence.
    If a word is not in the model's vocabulary, it is ignored.
    """
    words_in_vocab = [word for word in sentence if word in model.wv]
    if not words_in_vocab:  # If none of the words in the sentence are in the vocabulary
        return np.zeros(vector_size)
    word_vectors = np.array([model.wv[word] for word in words_in_vocab])
    return np.mean(word_vectors, axis=0)

# Step 4: Apply to All Sentences
vector_size = 100  # Same as the vector size of the model
avg_word2vec_features = np.array([
    compute_avg_word2vec(sentence, model, vector_size) for sentence in sentences
])

# Step 5: Output the Results
print("Average Word2Vec Features for Each Sentence:")
print(avg_word2vec_features)
```

```python
from tqdm import tqdm
```

```python

# Independent Feature (X)
X = []  # Store average Word2Vec vectors for each sentence
for i in tqdm(range(len(sentences))):  # Loop through tokenized sentences
    X.append(compute_avg_word2vec(sentences[i], model, vector_size))

X = np.array(X) 
```
